```python
from pathlib import Path
import pandas as pd
```

```python
# get the current directory
current_dir = Path.cwd()

# path to the processed data folder
processed_data_folder = current_dir.parents[1] / "data" / "processed"
```

```python
# load the combined features dataframe
df_final = pd.read_pickle(processed_data_folder / "binding_region_combined_features_expanded.pkl")

# view the first 4 rows
print(df_final.head())

# the columns
print(df_final.columns)
```

```python
# add length of sequence as a column
df_final["seq_length"] = df_final["binding_region_seq"].str.len()

# view the modified df
print(df_final.head())
```

```python
df_final.groupby(["binding_mode"]).agg(sequence_count = ('binding_region_seq', 'count'),
                                       mean_length = ('seq_length', 'mean'),
                                       sd_length = ('seq_length', 'std'),
                                       max_length = ('seq_length', 'max'),
                                       min_length = ('seq_length', 'min'))
```

```python
df_final.groupby(["binding_mode"]).agg(mean_relasa = ('avg_relasa', 'mean'),
                                       sd_relasa = ('avg_relasa', 'std'),
                                       mean_plddt = ('avg_plddt', 'mean'),
                                       sd_plddt = ('avg_plddt', 'std'),
                                       mean_fcr = ('fcr', 'mean'),
                                       sd_fcr = ('fcr', 'std'),
                                       mean_scd = ('scd', 'mean'),
                                       sd_scd = ('scd', 'std'),
                                       mean_shd = ('shd', 'mean'),
                                       sd_shd = ('shd', 'std'))
```

```python
df_final.groupby(["binding_mode"]).agg(mean_ss_H = ('ss3_freq_H', 'mean'),
                                       sd_ss_H = ('ss3_freq_H', 'std'),
                                       mean_ss_E = ('ss3_freq_E', 'mean'),
                                       sd_ss_E = ('ss3_freq_E', 'std'),
                                       mean_ss_C = ('ss3_freq_C', 'mean'),
                                       sd_ss_C = ('ss3_freq_C', 'std'))
```

```python
# 1. Define all 20 standard amino acid single-letter codes
sec_str = ["B", "E", "G", "H", "I", "P", "S", "T", "C"]

# 2. Dynamically build the aggregation dictionary
agg_dict = {}
for ss in sec_str:
    agg_dict[f'mean_ss_{ss}'] = (f'ss_freq_{ss}', 'mean')
    agg_dict[f'sd_ss_{ss}'] = (f'ss_freq_{ss}', 'std')

# 3. Pass the dictionary unpacker (**) into the .agg() method
df_final.groupby(["binding_mode"]).agg(**agg_dict)
```

```python
# 1. Define all 20 standard amino acid single-letter codes
amino_acids = ["A", "C", "D", "E", "F", "G", "H", "I", "K", "L", 
               "M", "N", "P", "Q", "R", "S", "T", "V", "W", "Y"]

# 2. Dynamically build the aggregation dictionary
agg_dict = {}
for aa in amino_acids:
    agg_dict[f'mean_aa_{aa}'] = (f'aa_freq_{aa}', 'mean')
    agg_dict[f'sd_aa_{aa}'] = (f'aa_freq_{aa}', 'std')

# 3. Pass the dictionary unpacker (**) into the .agg() method
df_final.groupby(["binding_mode"]).agg(**agg_dict)
```
