```python
import alphasync_test
from process_mobidb_scores import get_mobidb_scores, features_binding_region_df
```

```python
input_fasta = "mobidb_search_2026-05-05T08-28-46.fasta"
score_patterns = ["derived-binding_mode_disorder_to_disorder-mobi"]
data = get_mobidb_scores(input_fasta, score_patterns)
cols_to_check = ["derived_binding_mode_disorder_to_disorder_mobi"]
filtered_data = data.dropna(subset=cols_to_check, thresh=1)
filtered_data = filtered_data.head(20)
filtered_data["pLDDT"] = filtered_data["uniprot_id"].apply(lambda x: alphasync_test.parse_alphasync_data(x, "plddt10", ""))
filtered_data["secondary_strc"] = filtered_data["uniprot_id"].apply(lambda x: alphasync_test.parse_alphasync_data(x, "sec", ""))

print(filtered_data)
```
